```python
import numpy as np          #importing numpy & pandas library
import pandas as pd
lst=[7, 2, 0, 3, 4, 2, 5, 0, 3, 4]      #How to count distance to the previous zero
df = pd.DataFrame({'X':lst})      #creating dataFrame of values in lst
df['Y']=0                #creating new column 'Y' in df
distance=-1             #initialising distance value as -1
for i in range(0,len(lst)):         #running loop for each row in df
    if df.loc[i]['X'] == 0:        #checking if value in column 'X' is zero then set distance to zero
        distance = 0       
    elif distance == -1:             #for initial case writing the index of value if no zero found yet
        df.loc[i]['Y'] = i+1
    else:                          #else increasing distance value and copying into the column 'Y'
        distance = distance+1
        df.loc[i]['Y'] = distance
df                 #printing the dataframe
```

```python
#Creating a DatetimeIndex that contains each business day of 2015  
dti = pd.date_range(start='2015-01-01', end='2015-12-31', freq='B') 
s = pd.Series(np.random.rand(len(dti)), index=dti)        #using it to index a Series of random numbers
o1 = s[dti.weekday == 2].sum()          #sum of the values in s for every Wednesday
print("The sum of the values in s for every Wednesday is {}".format(o1),end='\n\n')

o2 = s.resample('M').mean()          #Average For each calendar month
print("Average For each calendar month is: ")
print(o2,end='\n\n')
o3 = s.groupby(pd.Grouper(freq='4M')).idxmax()         #The date on which the highest value occurred
print("The date on which the highest value occurred :")
print(o3)
```
